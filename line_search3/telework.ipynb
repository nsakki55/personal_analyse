```python
from japanmap import picture
from japanmap import pref_names
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
data = pd.read_excel('telework.xlsx', names=['area', 'No1', 'No2', 'No3'])
data.sort_values(by='No3')
```

```python
plt.figure(figsize=(10, 8))
plt.rcParams['font.size'] = 14
cmap = plt.get_cmap('Greens')
norm = plt.Normalize(vmin = data['No3'].min(), vmax=data['No3'].max())
fcol = lambda x: '#' + bytes(cmap(norm(x), bytes=True)[:3]).hex()
show_data = {}
for prefecture, c in zip(data['area'].values, data['No3'].apply(fcol).values):
    show_data[prefecture] = c

cbar = plt.colorbar(plt.cm.ScalarMappable(norm, cmap))
cbar.set_label('テレワーク実施割合 [%]', rotation=-90, fontsize=15,labelpad=20)
plt.title('オフィスワーク回答者のテレワーク実施割合 (第３回の全国調査)', fontsize=18, pad=10)
plt.tick_params(labelbottom=False, labelleft=False, bottom=False, left=False)
plt.tight_layout()
plt.imshow(picture(show_data))
plt.savefig('LINE_research3.png', dpi=300)
plt.show()
```
